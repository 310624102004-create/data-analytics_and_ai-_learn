```python

score = np.loadtxt('/content/survey.txt', dtype=int)
pass_count = sum(score >= 9)
print(f"Pass count (scores >= 9): {pass_count}")
fail_count = sum(score <= 6)
print(f"Fail count (scores <= 6): {fail_count}")
neutral_count = len(score) - pass_count - fail_count
print(f"Neutral count (scores > 6 and < 9): {neutral_count}")
```

```python
import pandas as pd

# Create a list of dictionaries for student data
student_data = [
    {'name': 'Alice', 'roll_number': 101},
    {'name': 'Bob', 'roll_number': 102},
    {'name': 'Charlie', 'roll_number': 103},
    {'name': 'David', 'roll_number': 104},
    {'name': 'Eve', 'roll_number': 105}
]

# Convert to a pandas DataFrame
df_students = pd.DataFrame(student_data)

# Save the DataFrame to a new CSV file
output_file_path = '/content/students.csv'
df_students.to_csv(output_file_path, index=False)

print(f"Student data saved to {output_file_path}")
print("Contents of students.csv:")
print(df_students.to_string(index=False))
```
