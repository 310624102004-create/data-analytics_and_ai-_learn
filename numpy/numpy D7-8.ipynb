
# Day 7 -Aggregate Functions in NumPy

```python
import numpy as np
arr = np.zeros((10,2)) # 2D array of 10 rows and 2 columns filled with zeros
print(arr)
```

```python
arr = np.zeros((2,10,2)) # 3D array of 2 blocks, each containing a 10x2 array filled with zeros
print(arr)
```

```python
arr = np.ones((10,2)) # 2D array of 10 rows and 2 columns filled with ones
print(arr)
```

```python
arr = np.ones((2,10,2)) # 3D array of 2 blocks, each containing a 10x2 array filled with ones
print(arr)
```

```python
arr = np.full((2,3), 7) # 2D array of 2 rows and 3 columns filled with the value 7
print(arr)
```

```python
arr=np.eye(2) # 2D identity matrix of size 2x2
print(arr,"\n")

arr=np.eye(3) # 3D identity matrix of size 3x3
print(arr)
```

```python
arr = np.empty((3,2,4)) # 2D array of 2 rows and 4 columns with uninitialized values
print(arr)
```

```python
arr = np.arange(0,100,2) # 1D array with values from 0 to 9 # starts at 0, ends before 100, with a step of 2
print(arr)
```

```python
arr = np.arange(0,100,1) # 1D array with values from 0 to 9
print(arr)
```

```python
arr = np.arange(0,100,15) # 1D array with values from 0 to 9
print(arr)
```

```python
arr = np.linspace(0,100,5) # start at 0, end at 100, with 5 (count number) evenly spaced values
print(arr)
```

```python
print(np.mean(arr)) # mean of the array 
print(np.median(arr)) # median of the array
print(np.std(arr)) # standard deviation of the array
print(np.var(arr)) # variance of the array
print(np.min(arr)) # minimum value in the array
print(np.sum(arr)) # sum of all elements in the array
print(np.max(arr)) # maximum value in the array
print(np.argmin(arr)) # index of the minimum value in the array
print(np.argmax(arr)) # index of the maximum value in the array
```

```python
arr=np.array([
              [1,2,3],
              [4,5,6]
            ]) # 2D array with specified values 
print(np.argmax(arr)) # index of the maximum value in the array (returns the index in the flattened array)
print(np.argmin(arr)) # index of the minimum value in the array (returns the index in the flattened array)
print(np.sum(arr,axis=0)) # sum of all elements in the array along axis 0 sums the columns
print(np.sum(arr,axis=1)) # sum of all elements in the array along axis 1 sums the rows
```


# Day 8  Filtering in NumPy, Where


```python
mark = np.array([50,100,65,45,76])
print(mark)
```

```python
print(mark[mark>50])
```

```python
print(mark>50)
```

```python
pass_mark = mark[mark>=50]
print(pass_mark)
```

```python
even = mark[mark%2==0]
print(even)
```

```python
ark = mark[(mark >= 50) & (mark <= 70)]
print(ark)
```

```python
ak = mark[(mark >= 50) | (mark%2==0)]
print(ak)
```

```python
# multi dimensional array filtering
arr = np.array([
              [1,2,3],
              [4,5,6]
            ]) # 2D array with specified values

print(arr)
```

```python
even = arr[arr%2==0] #this using flattened array to filter even numbers
print(even)
```

```python
even = np.where(arr%2==0,arr,0) # this using np.where to filter even numbers and replace odd numbers with 0 and shape of the original array is preserved
print(even)
```
